```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
cases = pd.read_csv('./data/covid.csv')
cases.head(1)
```

```python
us = cases[cases['countriesAndTerritories'] == 'United_States_of_America']
x = [[int(i.replace('2020-', ''))] for i in us['year_week'].values]
y = us['cases_weekly'].values
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=.8, test_size=.2, random_state=100)
plt.scatter(x_train, y_train, color='yellow')
plt.xlabel('Week')
plt.ylabel('Cases')
plt.title('Weekly COVID-19 Cases')
plt.show()
```

# Data Visualization

## Weekly COVID Cases

This data shows:

1. COVID is still very much a problem and it continues to worsen every day.
